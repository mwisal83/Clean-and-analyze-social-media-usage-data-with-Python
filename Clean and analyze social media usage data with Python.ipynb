## Analyzing Individual Feature Patterns using Visualization


### How to choose the right visualization method?
When visualizing individual variables, it is important to first understand what type of variable you are dealing with. This will help us find the right visualization method for that variable.



```python
#plt.figure(figsize=(30,20))
plt.title("Daily organic reach over time")
sns.scatterplot(x = df['Date'], y = df['Daily Organic Reach'])
```

```python
plt.figure(figsize=(20,20))
plt.title("Daily organic reach over a month")
sns.barplot(x = df['Daily Organic Reach'],y = df.index)
plt.xlabel("Number of days")
plt.ylabel("Organic reach")
```

```python
plt.figure(figsize=(20,20))
plt.title("Total Monthly Reach")
sns.barplot(x = df['Total_Monthly_Reach'],y = df.index)
plt.xlabel("Number of days")
plt.ylabel("Monthly reach")
```
